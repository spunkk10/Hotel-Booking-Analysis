```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# loading the dataset

```python
df=pd.read_csv("hotel_booking.csv",encoding="unicode_escape")
```

# Data Cleaning

```python
df.head()
```

```python
df.drop(["name","email","phone-number","credit_card"],axis=1,inplace=True)
```

```python
df.shape
```

```python
df.info()
```

```python
df["reservation_status_date"]=pd.to_datetime(df["reservation_status_date"])
```

```python
df.describe(include='object')
```

```python
for col in df.describe(include='object'):
    print(col)
    print(df[col].unique())
    print("-"*50)
```

```python
df.isnull().sum()
```

```python
df.drop(["agent","company"],axis=1,inplace=True)
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df=df[df["adr"]<5000]
```

```python
df.describe()
```

# Data Analysis and Visualization

```python
can= df["is_canceled"].value_counts(normalize=True)
print(can)
a=sns.countplot(x="is_canceled",data=df,hue="is_canceled")
plt.title("Reservation Status Count")
plt.figure(figsize=(5,5))
for bars in a.containers:
    a.bar_label(bars)
plt.show()
```

```python
a=sns.countplot(x="hotel",data=df,hue="is_canceled")
plt.title("Reservation Status Count")
plt.figure(figsize=(5,5))
for bars in a.containers:
    a.bar_label(bars)
plt.show()
```

```python
print(df["hotel"].value_counts())
```

```python
resort_hotel=df[df["hotel"]== "Resort Hotel"]
print("Resort Hotel")
print(resort_hotel["is_canceled"].value_counts(normalize=True))
print("-"*10)
city_hotel=df[df["hotel"]== "City Hotel"]
print("City Hotel")
print(city_hotel["is_canceled"].value_counts(normalize=True))
```

```python
resort_hotel=resort_hotel.groupby(["reservation_status_date"],as_index=False)["adr"].mean()
city_hotel=city_hotel.groupby(["reservation_status_date"],as_index=False)["adr"].mean()
```

```python
plt.figure(figsize=(20,8))
sns.lineplot(x="reservation_status_date",y="adr",data=resort_hotel,label="Resort Hotel")
sns.lineplot(x="reservation_status_date",y="adr",data=city_hotel,label="City Hotel")
plt.title("Average daily rate in City and Resort hotel")
plt.legend(fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(15,5))
df["month"]=df["reservation_status_date"].dt.month
ax=sns.countplot(x="month",hue="is_canceled",data=df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Reservation Status per month")
plt.legend(["not cancelled","cancelled"])
plt.show()
```

```python
resort_hotell=df[df["hotel"]== "Resort Hotel"]
plt.figure(figsize=(15,5))
ax=sns.countplot(x="month",hue=resort_hotell["is_canceled"],data=resort_hotell)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Reservation Status per month of Resort hotel")
plt.legend(["not cancelled","cancelled"])
plt.show()
```

```python
city_hotell=df[df["hotel"]== "City Hotel"]
plt.figure(figsize=(15,5))
ax=sns.countplot(x="month",hue=city_hotell["is_canceled"],data=city_hotell)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Reservation Status per month of City hotel")
plt.legend(["not cancelled","cancelled"])
plt.show()
```

```python
plt.figure(figsize=(20,8))
plt.title("Average daily rate per month")
sns.barplot(x="month",y="adr",data=df[df["is_canceled"]==1].groupby("month",as_index=False)["adr"].sum(),hue="month",palette="Accent")
plt.show()
```

```python
cancelled_data=df[df["is_canceled"]==1]
top10cont= cancelled_data["country"].value_counts()[:10]
plt.figure(figsize=(5,5))
plt.pie(top10cont,autopct="%.2f",labels=top10cont.index)
plt.title("top 10 country with cancelllation")
plt.show()
```

```python
cancelled_df_adr=cancelled_data.groupby(["reservation_status_date"],as_index=False)["adr"].mean().sort_values(by="reservation_status_date")

not_cancelled_data=df[df["is_canceled"]==0]
not_cancelled_df_adr=not_cancelled_data.groupby(["reservation_status_date"],as_index=False)["adr"].mean().sort_values(by="reservation_status_date")

plt.figure(figsize=(20,8))
plt.title("Average Daily Rate")
plt.plot(not_cancelled_df_adr["reservation_status_date"],not_cancelled_df_adr["adr"],label="Not Cancelled")
plt.plot(cancelled_df_adr["reservation_status_date"],cancelled_df_adr["adr"],label="Cancelled")
plt.legend()
plt.show()
```
